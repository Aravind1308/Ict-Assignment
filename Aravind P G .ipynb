### Visualizations using the dataset

```python
plt.hist(data=iris_data ,x='Classification')
plt.title('Count of classification column')
plt.show()
```

###### > Count of each set = 50

### Comparing Sepal Length and Sepal Width

```python

sns.scatterplot(data=iris_data,x='SL', y='SW',hue='Classification')
plt.title('Comparing Sepal Length and Sepal Width')
plt.show()
```

###### > Iris-Setosa has smaller sepal lengths but larger sepal widths.
###### > Iris-Versicolor lies in the middle of the other two in terms of sepal length and width.
###### > Iris-Virginica has larger sepal lengths but smaller sepal widths.

### Comparing Petal Length and Petal Width

```python
sns.scatterplot(data=iris_data,x='PL', y='PW',hue='Classification')
plt.title('Comparing Petal Length and Petal Width')
plt.show()
```

###### > Iris-Setosa has smaller petal lengths and widths.
###### > Iris-Versicolor lies in the middle of the other two in terms of petal length and width.
###### > Iris-Virginica has the largest of petal lengths and widths.

### Correlation between variables

```python
sns.heatmap(iris_data.corr(),annot= True)
plt.title('Correlation between variables')
plt.show()
```

##### > Petal width and Petal length have high correlations. 
##### > Petal length and Sepal width have low correlations.
##### > Petal Width and Sepal length have good correlations.
